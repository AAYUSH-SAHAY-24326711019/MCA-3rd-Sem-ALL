```python
# Area of triangle program
import math
a = float(input("Input the value of 1st side"))
b = float(input("Input the value of 2nd side"))
c = float(input("Input the value of 3rd side"))
s = (a+b+c)/2

ar = math.sqrt(s*(s-a)*(s-b)*(s-c))
print("The area is : ",ar)
```
