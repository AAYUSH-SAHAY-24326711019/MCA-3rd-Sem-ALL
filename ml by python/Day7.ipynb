```python
a=set()
b=set()

for i in range (1,3):
    a.add(i)

for i in range (3,5):
    b.add(i)

print("Currently a set is :",a)

print("Currently b set is :",b)

print("result of a.isdisjoint(b)  ",a.isdisjoint(b))
```

```python
print(all(a))   #if all the elements of a are non zero

#suppose
x=set()
x.add(0)
x.add(1)

print(all(x)) # returns false Since x has an element 0

print(any(x)) # returns true if any value inside set is 0
```

```python
#there is a function enumerate
#it has:
'''consider there is a set , theoretically it has an index, with each index , make a tuple
    of all the elements with its index.

    if a=(1,2)
    we want(0,1),(1,2) .... (index,value) just like a dictionary and not a dictionary directly.
'''
# x has {0,1}
# enumerate will mix and not print, so we will be using the for loop with it.
for i in enumerate(x):
    print(i)
```
